Question 3: The ice-cream shop has also aggregated their sales data by month and ice-cream flavor for the previous year. 

```python
flav_df = pd.read_csv('Monthly Ice Cream Shop Sales by Flavor.csv')
flav_df['Month Name'] = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
flav_df.head()
```

A. Create a line graph that displays the total units across all months (time on x axis). Make sure you have the appropriate labels for the x and y axis.

```python
fig, ax = plt.subplots()
ax.plot(flav_df['Month Name'], flav_df['Total Units'], label='Total Units Sold')
ax.set_xlabel('Month')
ax.set_ylabel('Units Sold')
ax.legend()
```

Optional: Create another graph with a separate line for each ice cream flavor to compare sales by flavors over time.

```python
fig, ax = plt.subplots()
tr = mtrans.offset_copy(ax.transData, fig=fig, x=0.0, y=-1.5, units='points')
ax.plot(flav_df['Month Name'], flav_df['Vanilla'], label='Vanilla', color='#EEDEC5')
ax.plot(flav_df['Month Name'], flav_df['Chocolate'], label='Chocolate', color='#724e2c')
ax.plot(flav_df['Month Name'], flav_df['Strawberry'], label='Strawberry', color='#FB6E7F')
ax.plot(flav_df['Month Name'], flav_df['Neapolitan'], label='Neapolitan', color='#3A9BDC')
ax.plot(flav_df['Month Name'], flav_df['Mango'], label='Mango', color='orange')
ax.plot(flav_df['Month Name'], flav_df['Lime'], label='Lime', color='#CCFF00', transform=tr)
ax.set_xlabel('Month')
ax.set_ylabel('Units Sold')
ax.legend()
```

B. Create a bar chart that displays the total profit per month. Make sure you have the appropriate labels for the x and y axis.

```python
sns.barplot(data=flav_df, x='Month Name', y='Total Profit')
```
